### ✳️ GitHub 가입 및 사용법
#### 1️⃣ GitHub 이해
- GitHub(깃허브)는 분산 버전 관리 시스템인 Git(깃)을 사용하는 프로젝트를 위한 웹 기반 호스팅 플랫폼이자, 개발자들의 소셜 네트워크 서비스이다.
- 2018년 마이크로소프트가 인수한 이곳은 전 세계 개발자들이 소스 코드를 공개(오픈소스)하고, 공동 작업(협업)하며 포트폴리오를 관리하는 공간으로 활용된다
- Github 기본 용어 이해
    - Repository : 저장소 / 말 그대로 파일이나 폴더를 저장해두는 저장소입니다.
    - Commit : 파일을 추가, 변경한 내용을 저장소에 저장하는 작업
    - push : 추가, 변경한 내용을 원격 저장소에 업로드하는 작업

#### 2️⃣ GitHub 회원가입 및 저장소 생성
1. https://github.com/ 접속후 우측 상단의 'Sign up'을 클릭하여 회원가입
2. 로그인 후 메인 페이지 왼쪽 상단의 'New' 버튼(또는 Create repository)을 클릭
3. 저장소 생성시 설정 사항
    - Repository name: 저장소 이름을 입력(필수)
    - Description: 저장소에 대한 설명(선택)
    - Public/Private: 공개(Public) 또는 비공개(Private) 중 선택
    - Add README : readme.md 파일 자동 생성으로 on 선택
    - 'Create repository' 버튼을 눌러 저장소를 만듬

    

4. 새로운 저장소(예: git_test)가 생성되어 readme.md 파일이 자동생성되어 화면에 표시

    

#### 3️⃣ GitHub 웹 브라우저에서 직접 파일 관리하기 
1. GitHub 웹 브라우저에서 직접 파일 관리하기 
    - 파일 업로드: Add file → Upload files 버튼을 클릭하여 드래그 앤 드롭으로 파일을 올림
    - 파일 생성 : Add file → Create new file을 눌러 새 파일 생성
    - 파일 수정 : 기존 파일을 클릭한 뒤 연필 아이콘(Edit this file)을 눌러 직접 내용을 수정
    - 변경사항 저장: 수정한 후 하단의 Commit changes 버튼을 눌러 저장
2. VS Code 웹 에디터 (github.dev)에서 파일 관리도 가능하다
    - GitHub 저장소 페이지에서 키보드의 . (마침표) 키를 누르면 브라우저 내에서 VS Code 환경이 열린다
    - 로컬 설치 없이 IDE 환경에서 파일을 수정하고, 왼쪽 소스 제어 탭에서 변경 사항을 커밋 및 푸시할 수 있다

### ✳️ GitHub Copilot 개요   
- GitHub Copilot은 GitHub가 OpenAI와 협업하여 2021년 출시한 자동 코드 완성 인공지능이다.
    - 2024년 12월 18일부터 무료 플랜을 제공하며, Visual Studio(2022 이후) 및, Visual Studio Code에도 기본 탑재되어 있다.
    - Visual Studio에서 사용시 코파일럿 채팅에서 에이전트로 설정하면 직접 프로젝트를 분석해 코드를 수정-빌드-재수정 과정을 반복한다.
    - https://github.com/features/copilot/ 

#### 1️⃣ GitHub Copilot 설치
1. VSCode에 Copilot 설치하기
    - VSCode 익스텐션 검색창에서 “copilot”을 입력하고 검색
    - 목록에서 ‘GitHub Copilot’과 ‘GitHub Copilot Chat’을 차례대로 설치

2. GitHub 계정 로그인 및 활성화
    - 우측하단에 로그인 알림이 뜸 > 로그인하여 GitHub 계정으로 인증 > 
    - "Authorize Visual Studio Code"를 클릭 > 연동이 완료되면 우측 하단 상태바에 코파일럿 아이콘이 활성화

    


#### 2️⃣ 무료 플랜에서 선택 가능한 주요 모델
- 2026년 기준, GitHub Copilot 무료 플랜(Free Tier)은 개인 개발자에게 매달 2,000개의 코드 자동 완성(Autocomplete)과 50회의 채팅 요청을 제공하며, 이 안에서 GPT-4o 및 Claude Haiku 4.5 같은 최상급 모델을 사용할 수 있다
- 사용 팁으로 복잡한 기획은 GPT-5나 Claude 4 같은 고성능 모델로 수행하고, 실제 구현 및 반복적인 코드 수정 단계에서 GPT-5 mini로 전환하면 무료 제공 횟수를 효율적으로 아낄 수 있다
